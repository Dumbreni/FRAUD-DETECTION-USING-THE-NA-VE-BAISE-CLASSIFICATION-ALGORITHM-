# Training split test { learn with 80% of data and excute what learnt with the remaining data} 

```python
from sklearn.model_selection import train_test_split
```

```python
y=target
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=100)
```

# telling/training the system how to allocate data, ie we are teaching the machine what to do with data

```python
from sklearn.naive_bayes import BernoulliNB
model=BernoulliNB()
```

```python
model.fit(x_train,y_train)
```

# Model Evaluation 

# 1. Testing Accuracy as a percentage

```python
accuracy=model.score(x_test,y_test)
```

```python
accuracy*100
```

# 2. Confusion Matrics

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
predictions=model.predict(x_test)
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
print(classification_report(y_test,predictions))
```
